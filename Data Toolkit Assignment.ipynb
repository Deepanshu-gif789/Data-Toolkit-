#**Data Toolkit Theoritical**

1. **What is NumPy, and why is it widely used in Python?**
      - NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides support for arrays, matrices, and a variety of mathematical functions. NumPy is widely used because it offers fast operations on large datasets through its powerful N-dimensional array object, efficient broadcasting, and vectorized operations, which are optimized in C under the hood.



2. **How does broadcasting work in NumPy?**
     - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. Instead of explicitly replicating smaller arrays, broadcasting stretches them to match the shape of the larger array. This saves memory and computation time and enables operations like adding a scalar to a matrix or combining arrays of different dimensions efficiently.

3. **What is a Pandas DataFrame?**
     - A Pandas DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns). It is similar to an Excel spreadsheet or a SQL table and allows for easy data manipulation, analysis, and cleaning.

4. **Explain the use of the groupby() method in Pandas.**
     - The groupby() method in Pandas is used to split data into groups based on certain criteria, perform operations like aggregation or transformation on these groups, and then combine the results. This is useful for analyzing data by categories (e.g., sales by region or gender-based statistics).



5. **Why is Seaborn preferred for statistical visualizations?**
     - Seaborn is preferred for statistical visualizations because it provides a high-level API for drawing attractive and informative plots. It is built on top of Matplotlib and integrates well with Pandas DataFrames. Seaborn simplifies the creation of complex plots like violin plots, boxplots, and heatmaps with minimal code.



6. ** What are the differences between NumPy arrays and Python lists?**

     - Memory Efficiency: NumPy arrays are more memory-efficient than Python lists.

     - Performance: NumPy operations are faster due to optimized C implementation.

     - Functionality: NumPy supports vectorized operations, broadcasting, and multidimensional arrays, which lists do not.

     - Homogeneity: NumPy arrays store elements of the same data type, while lists can contain different data types.

7. **What is a heatmap, and when should it be used?**
     - A heatmap is a graphical representation of data where individual values are represented as colors. It is commonly used to show correlations between variables, matrix data, or the frequency of events in a dataset. Heatmaps are especially useful for visualizing large datasets and spotting patterns.

8. **What does the term “vectorized operation” mean in NumPy?**
     - A vectorized operation in NumPy means applying an operation to an entire array without the need for explicit loops. These operations are faster and more concise, taking advantage of low-level optimizations in NumPy.



9. **How does Matplotlib differ from Plotly?**
     - Matplotlib: A 2D plotting library that produces static, publication-quality visualizations.

     - Plotly: An interactive plotting library that enables web-based, dynamic, and zoomable visualizations.
     - Plotly is generally preferred for dashboards and interactive applications, while Matplotlib is better suited for static reports.

10. **What is the significance of hierarchical indexing in Pandas?**
     - Hierarchical indexing (MultiIndex) allows users to work with higher-dimensional data in a lower-dimensional form. It enables complex data manipulation, such as reshaping and drilling down into data, making it useful for time series or panel data.



11. **What is the role of Seaborn’s pairplot() function?**
     - The pairplot() function creates a grid of scatter plots for each pair of variables in a dataset, along with histograms or KDE plots on the diagonal. It is useful for exploring relationships between multiple numerical features.

12. **What is the purpose of the describe() function in Pandas?**
     - The describe() function provides summary statistics of a DataFrame or Series, including measures like mean, median, standard deviation, min, and max. It offers a quick overview of the distribution and central tendency of the data.



13. **Why is handling missing data important in Pandas?**
     - Missing data can distort statistical analysis and machine learning models. Pandas provides methods like dropna(), fillna(), and interpolation to handle missing values. Proper handling ensures data integrity and accurate results.



14. **What are the benefits of using Plotly for data visualization?**
     - Interactive plots (zoom, pan, hover)

     - Web-based and shareable

     - Supports a wide variety of chart types (3D, maps, etc.)

     - Integrates with Dash for dashboard creation



15. **How does NumPy handle multidimensional arrays?**
      
    - NumPy handles multidimensional arrays using its ndarray structure. It supports operations across dimensions, reshaping, slicing, and broadcasting for efficient numerical computations on data from 1D to nD.

16. **What is the role of Bokeh in data visualization?**
     - Bokeh is a Python library for interactive and real-time visualizations. It provides high-performance interactivity for large datasets and is well-suited for creating dashboards and web applications.

17. **Explain the difference between apply() and map() in Pandas.**

     - map() is used for element-wise transformations on Series.

     - apply() is used on both Series and DataFrames and can apply functions across rows or columns, offering more flexibility.



18. **What are some advanced features of NumPy?**

     - Broadcasting and vectorized operations

     - Support for complex numbers

     - Linear algebra and Fourier transform capabilities

     - Memory mapping of large files

     - Integration with C/C++ and Fortran



19. **How does Pandas simplify time series analysis?**
     - Pandas provides dedicated datetime functions, resampling tools, rolling window statistics, and time-based indexing. This allows for easy handling of time series data for forecasting, trend analysis, and date manipulation.

20. **What is the role of a pivot table in Pandas?**
     - A pivot table reorganizes data by aggregating values using a specified function (mean, sum, etc.) across one or more keys. It is useful for summarizing large datasets and drawing insights.



21. **Why is NumPy’s array slicing faster than Python’s list slicing?**
     - NumPy uses contiguous memory blocks and vectorized operations in C, which are much faster than Python lists that are dynamic and use pointers. NumPy avoids the overhead of type checking and memory allocation.

22. **What are some common use cases for Seaborn?**

     - Visualizing relationships (scatter plots, pairplots)

     - Statistical analysis (violin, box, and bar plots)

     - Distribution analysis (histograms, KDE)

     - Correlation heatmaps

     - Data exploration in data science pipelines



# **Practical Question**

1. **Create a 2D NumPy array and calculate the sum of each row.**

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("Row sums:", row_sum)
```

2. **Write a Pandas script to find the mean of a specific column in a DataFrame.**

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
mean_value = df['B'].mean()
print("Mean of column B:", mean_value)
```

3. **Create a scatter plot using Matplotlib.**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

4. **How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.**

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
correlation = df.corr()
sns.heatmap(correlation, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

5. **Generate a bar plot using Plotly.**

```python
import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Orange'], 'Count': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Fruits', y='Count', title='Fruit Count')
fig.show()
```

6. **Create a DataFrame and add a new column based on an existing column.**

7. **Write a program to perform element-wise multiplication of two NumPy arrays?**

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise multiplication:", result)
```

8. **Create a line plot with multiple lines using Matplotlib.**

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 5, 10, 17]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title("Multiple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})
fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
